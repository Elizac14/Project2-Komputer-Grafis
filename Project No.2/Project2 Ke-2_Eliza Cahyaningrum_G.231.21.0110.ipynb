```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Membaca file CSV
data = pd.read_csv('suhu permukaan.csv')
```

```python
data
```

```python
# Membuat DataFrame dari dataset
data = {
    'Tanggal': ['24 Juli 2013', '24 Juli 2013', '24 Juli 2013', '24 Juli 2013',
                '24 Juli 2013', '24 Juli 2013', '24 Juli 2013', '24 Juli 2013',
                '25 Juli 2013'],
    'Waktu': ['24 Juli 2013 00.00 UTC', '24 Juli 2013 03.00 UTC', '24 Juli 2013 06.00 UTC', '24 Juli 2013 09.00 UTC',
              '24 Juli 2013 12.00 UTC', '24 Juli 2013 15.00 UTC', '24 Juli 2013 18.00 UTC', '24 Juli 2013 21.00 UTC',
              '25 Juli 2013 00.00 UTC'],
    'Asimilasi': [23.3, 25.5, 25.6, 24.8, 22.5, 22.4, 22.95, 22.0, 25.8],
    'Observasi': [24.2, 25.2, 23.9, 23.9, 24.0, 24.2, 24.0, 22.8, 23.0],
    'Non Asimilasi': [23.6, 25.7, 25.3, 25.5, 24.9, 25.1, 24.7, 24.3, 25.9]
}

df = pd.DataFrame(data)

# Membuat plot menggunakan Seaborn
plt.figure(figsize=(8, 5))
sns.lineplot(data=df, x='Waktu', y='Asimilasi', marker='o', label='Asimilasi', color='blue')
sns.lineplot(data=df, x='Waktu', y='Observasi', marker='o', label='Observasi', color='yellow')
sns.lineplot(data=df, x='Waktu', y='Non Asimilasi', marker='o', label='Non Asimilasi', color='red')

# Mengatur rentang sumbu y
plt.ylim(20, 27)

# Menghilangkan koma pada sumbu y
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: int(x)))

# Menambahkan judul dan label sumbu pada plot
plt.xlabel('Waktu\n(Tanggal)', weight='bold')
plt.ylabel('Celcius', weight='bold')
plt.title('Suhu Permukaan', weight='bold')

# Memutar label sumbu x agar tidak bertumpuk
plt.xticks(rotation=45)

# Menampilkan legenda
plt.legend()

# Menampilkan plot
plt.tight_layout()
plt.show()
```
